```python
import csv 
import pandas as pd 
import numpy as np 
import mediapipe as mp 
import cv2

```

```python
rock = pd.read_csv('rock.csv')
paper = pd.read_csv('paper.csv')
scissor = pd.read_csv('scissors.csv')
```

```python
rock.columns
```

```python
rock['label'] = pd.Series(np.ones(len(rock)))
paper['label'] = pd.Series(np.ones(len(paper))*2)
scissor['label'] = pd.Series(np.ones(len(scissor))*3)
rock.head(), paper.head(), scissor.head()
```

```python
rock = rock.drop(['Unnamed: 0'], axis=1)
paper = paper.drop(['Unnamed: 0'], axis=1)
scissor = scissor.drop(['Unnamed: 0'], axis=1)

rock.head()
```

```python
rock = rock[1:]
paper = paper[1:]
scissor = scissor[1:]

rock.head(), paper.head(), scissor.head()
```

```python
rock.shape, paper.shape, scissor.shape
```

```python
data = pd.concat([rock,paper,scissor], axis=0)
data.head
```

```python
data = data.sample(frac=1).reset_index(drop=True)
data.head()
```

```python
data.label.unique()
```

```python
label = data.label.copy()
```

```python
label.head()
```

```python
label = label.astype('int')
label.head()
```

```python
data.drop(['label'], axis=1,inplace=True)
data.head()
```

```python
np.random.seed(225)
mask = np.random.rand(len(data)) < 0.70
train_data = data[mask]
test_data = data[~mask]

train_label = label[mask]
test_label = label[~mask]
```

```python
train_data.shape, test_data.shape, test_label.shape, train_label.shape
```

```python
train_data.head(20)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, Sequential, optimizers
from tensorflow.keras.utils import to_categorical
```

```python
train_label = to_categorical(train_label)
test_label = to_categorical(test_label)
```

```python
model = models.Sequential()
model.add(layers.Dense(24, activation='relu',  input_shape=(train_data.shape[1],)))
model.add(layers.Dense(24, activation='relu'))
model.add(layers.Dense(4))
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_data, train_label, epochs=50, batch_size=100)
```
